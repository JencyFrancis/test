```python
# Statistical steps for feature importance

# Initialize results list
results = []

# Chi-square tests (Type vs flags)
for flag in ['Machine_failure', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF']:
    chi2, p = chi2_contingency(pd.crosstab(df1_scaled['Type_encoded'], df1_scaled[flag]))[:2]
    results.append({
        'Important Feature': flag,
        'Result': 'Significant association' if p < 0.05 else 'No association',
        'Test Type': 'Chi-square',
        'Test Statistic': f"{chi2:.3f}",
        'p-value': f"{p:.4f}"
    })

# T-tests (Numerical features vs failure)
for num_var in ['Air_temperature', 'Process_temperature', 'Rotational_speed', 'Torque', 'Tool_wear']:
    t_stat, p = ttest_ind(
        df1_scaled[df1_scaled['Machine_failure'] == 1][num_var],
        df1_scaled[df1_scaled['Machine_failure'] == 0][num_var],
        equal_var=False
    )
    results.append({
        'Important Feature': num_var,
        'Result': 'Significant difference' if p < 0.05 else 'No difference',
        'Test Type': 'T-test',
        'Test Statistic': f"{t_stat:.3f}",
        'p-value': f"{p:.4f}"
    })

# Create styled DataFrame
styled_df = (
    pd.DataFrame(results)
    .style
    .hide(axis="index")
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([
        {'selector': 'th, td', 'props': 'border: 1px solid black;'},
        {'selector': 'th', 'props': 'background-color: #f7f7f7;'}
    ])
)

# Display table
display(styled_df)
```

## Oversampling Technique

```python
# Initialize storage for oversampled data
oversampled_data = {}

# Define oversampling techniques
oversampling_techniques = {
    'None': None,
    'SMOTE': SMOTE(random_state = 17),
    'ADASYN': ADASYN(random_state = 17),
    'RandomOverSampler': RandomOverSampler(random_state = 17),
    'SMOTETomek': SMOTETomek(random_state = 17)
}

# List to store summary info
summary = []

# Apply oversampling techniques
for name, technique in oversampling_techniques.items():
    if technique is None:
        X_res, y_res = X_train, y_train
    else:
        try:
            X_res, y_res = technique.fit_resample(X_train, y_train)
        except Exception as e:
            print(f"Error with {name}: {e}")
            X_res, y_res = X_train, y_train

    # Store results
    oversampled_data[name] = (X_res, y_res)

    # Calculate failure count
    failure_count = sum(y_res)

    # Calculate failure rate
    failure_rate = round(y_res.mean(), 4)

    # Add results to summary
    summary.append({
        'Technique': name,
        'Original Samples': len(y_train),
        'Resampled Samples': len(y_res),
        'Failure Count': failure_count,
        'Failure Rate': round(y_res.mean(), 4)
    })

# Convert results to a table
summary_df = pd.DataFrame(summary)

#Print the summary table
print("\nOversampling Summary:\n")
print(summary_df.to_string(index = False))
```

```python
#FAILURE DISTRIBUTION AFTER OVERSAMPLING

# Create failure_types_oversampled dataframe
failure_types_oversampled = pd.DataFrame({
    'Machine_failure': y_train,
    'Failure_type': df1.loc[X_train.index, 'Failure_type']
})

# Prepare data
failure_categories = ['Power Failure', 'Overstrain Failure',
                     'Tool Wear Failure', 'Heat Dissipation Failure']

category_colors = {'Power Failure': 'gold',
                  'Overstrain Failure': 'violet',
                  'Tool Wear Failure': 'lime',
                  'Heat Dissipation Failure': 'orange'}

# Get original failure type distribution from training data
original_failure_dist = failure_types_oversampled[failure_types_oversampled['Machine_failure'] == 1]['Failure_type'].value_counts(normalize = True)

# Create plot with correct technique order
fig, axes = plt.subplots(2, 2, figsize = (12, 10))
techniques = ['SMOTE', 'ADASYN', 'RandomOverSampler', 'SMOTETomek']

# Get the total failure count from summary table
for ax, technique in zip(axes.flatten(), techniques):
    summary_row = summary_df[summary_df['Technique']==technique].iloc[0]
    total_failures = summary_row['Failure Count']

    # Calculate expected counts for each failure type
    failure_counts = {}
    for category in failure_categories:
        if technique == 'None':
            # For original data, use actual counts
            failure_counts[category] = failure_types_oversampled[
                (failure_types_oversampled['Machine_failure']==1) &
                (failure_types_oversampled['Failure_type']==category)
            ].shape[0]
        else:
            # For oversampled data, maintain original proportions
            failure_counts[category] = round(original_failure_dist.get(category, 0) * total_failures)

    # Adjust for any rounding errors
    total = sum(failure_counts.values())
    if total != total_failures:
        difference = total_failures - total
        # Add the difference to the largest category
        largest_cat = max(failure_counts, key=failure_counts.get)
        failure_counts[largest_cat] += difference

    # Create plot with consistent colors
    counts = [failure_counts[cat] for cat in failure_categories]
    ax.pie(counts,
           labels = failure_categories,
           autopct = lambda p: f'{p:.2f}%' if p > 0 else '',
           colors = [category_colors[cat] for cat in failure_categories],
           wedgeprops = {'linewidth': 1, 'edgecolor': 'white'})
    # Set main title
    plt.suptitle('Failure Distribution after Oversampling', fontsize = 16, fontweight = 'bold')

    # Set title
    ax.set_title(f"{technique}\n(Total failures: {total_failures:,})")

# Adjust layout
plt.tight_layout()

# Display plot
plt.show()
```
