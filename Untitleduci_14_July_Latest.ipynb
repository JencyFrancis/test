```python
# Statistical steps for feature importance

# Initialize results list
results = []

# Chi-square tests (Type vs flags)
for flag in ['Machine_failure', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF']:
    chi2, p = chi2_contingency(pd.crosstab(df1_scaled['Type_encoded'], df1_scaled[flag]))[:2]
    results.append({
        'Important Feature': flag,
        'Result': 'Significant association' if p < 0.05 else 'No association',
        'Test Type': 'Chi-square',
        'Test Statistic': f"{chi2:.3f}",
        'p-value': f"{p:.4f}"
    })

# T-tests (Numerical features vs failure)
for num_var in ['Air_temperature', 'Process_temperature', 'Rotational_speed', 'Torque', 'Tool_wear']:
    t_stat, p = ttest_ind(
        df1_scaled[df1_scaled['Machine_failure'] == 1][num_var],
        df1_scaled[df1_scaled['Machine_failure'] == 0][num_var],
        equal_var=False
    )
    results.append({
        'Important Feature': num_var,
        'Result': 'Significant difference' if p < 0.05 else 'No difference',
        'Test Type': 'T-test',
        'Test Statistic': f"{t_stat:.3f}",
        'p-value': f"{p:.4f}"
    })

# Create styled DataFrame
styled_df = (
    pd.DataFrame(results)
    .style
    .hide(axis="index")
    .set_properties(**{'text-align': 'center'})  # Center-align all cells
    .set_table_styles([  # Add borders
        {'selector': 'th, td', 'props': 'border: 1px solid black;'},
        {'selector': 'th', 'props': 'background-color: #f7f7f7;'}
    ])
)

# Display final table
display(styled_df)
```

```python
# Select features for modeling
feature_columns = [
    'Air_temperature', 'Process_temperature', 'Rotational_speed','Torque', 'Tool_wear', 'Type_encoded']

# Define Feature Matrix(X) and Target(Y)
X = df1_scaled[feature_columns]
y = df1_scaled['Machine_failure']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=17, stratify=y
)

# Print the sizes and distribution
print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
print(f"Training set failure rate: {y_train.mean():.4f}")
print(f"Test set failure rate: {y_test.mean():.4f}")

# Convert to NumPy arrays
X_train_scaled = X_train.values
X_test_scaled = X_test.values
```

## Oversampling Technique

```python
# List to store summary info
summary = []

# Apply oversampling techniques
for name, technique in oversampling_techniques.items():
    if technique is None:
        X_res, y_res = X_train, y_train
    else:
        try:
            X_res, y_res = technique.fit_resample(X_train, y_train)
        except Exception as e:
            print(f"Error with {name}: {e}")
            X_res, y_res = X_train, y_train

    # Store results
    oversampled_data[name] = (X_res, y_res)
    summary.append({
        'Technique': name,
        'Original Samples': len(y_train),
        'Resampled Samples': len(y_res),
        'Failure Rate': round(y_res.mean(), 4)
    })

# Convert results to a table
summary_df = pd.DataFrame(summary)
print("\nOversampling Summary:\n")
print(summary_df.to_string(index=False))
```

## Model Definition and Training

```python
# Define models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=17, n_jobs=-1),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=17),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

# Results storage
results = []
```

```python
# Store results for all models and sampling methods
results = []

# Train and evaluate models
for model_name, model in models.items():
    for sampling_name, (X_train_resampled, y_train_resampled) in oversampled_data.items():
        try:
            # Train the model on resampled data
            model.fit(X_train_resampled, y_train_resampled)

            # Predict on the test set (df1_scaled is already scaled)
            y_pred = model.predict(X_test)
            y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

            # Calculate metrics
            results.append({
                'Model': model_name,
                'Sampling': sampling_name,
                'Accuracy': accuracy_score(y_test, y_pred),
                'Precision': precision_score(y_test, y_pred),
                'Recall': recall_score(y_test, y_pred),
                'F1_Score': f1_score(y_test, y_pred),
                'MCC': matthews_corrcoef(y_test, y_pred),
                'ROC_AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else np.nan
            })

        except Exception as e:
            print(f"{model_name} with {sampling_name}: Error - {e}")

# Convert results to DataFrame and display as table
results_df = pd.DataFrame(results)
print("\nModel Evaluation Results:\n")
print(results_df.round(3).to_string(index=False))
```

```python
# Define parameter grids for each model
param_grids = {
    'Random Forest': {
        'n_estimators': [50, 100, 200],             # Number of trees in the forest
        'max_depth': [10, 20, None],                # Maximum depth of the tree
        'min_samples_split': [2, 5],                # Minimum number of samples required to split an internal node
        'min_samples_leaf': [1, 2]                  # Minimum number of samples required to be at a leaf node
    },
    'Neural Network': {
        'hidden_layer_sizes': [(50,), (100,), (100, 50), (100, 50, 25)],  # Hidden layer architectures
        'alpha': [0.0001, 0.001, 0.01],             # L2 regularization strength
        'learning_rate_init': [0.001, 0.01]         # Initial learning rate
    },
    'KNN': {
        'n_neighbors': [3, 5, 7, 9],                # Number of neighbors to use
        'weights': ['uniform', 'distance'],         # Weight function used in prediction
        'metric': ['euclidean', 'manhattan']        # Distance metric
    }
}

# Select best sampling technique based on F1 Score
best_sampling = performance_results.loc[performance_results['F1_Score'].idxmax(), 'Sampling']  # Get sampling technique with highest F1
best_X_train, best_y_train = oversampled_data[best_sampling]  # Retrieve the corresponding resampled data

print(f"Using best sampling technique: {best_sampling}")

# Initialize a list to store optimized model results
optimized_results = []

# Use Stratified K-Fold cross-validation (preserves class distribution)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)

# Loop over each model to perform grid search optimization
for model_name, model in models.items():
    print(f"\nOptimizing {model_name}...")

    # Set up GridSearchCV with scoring based on F1 score
    grid_search = GridSearchCV(
        model,  # Model instance
        param_grids[model_name],  # Corresponding hyperparameter grid
        cv=cv,  # Stratified 5-fold cross-validation
        scoring='f1',  # Use F1 score for evaluation
        n_jobs=-1,  # Use all available CPU cores
        verbose=0  # No verbosity in output
    )

    try:
        # Fit the grid search on training data
        grid_search.fit(best_X_train, best_y_train)

        # Retrieve the best performing model from the grid search
        best_model = grid_search.best_estimator_

        # Make predictions on test data
        y_pred = best_model.predict(X_test)

        # Predict probabilities (if supported)
        y_pred_proba = best_model.predict_proba(X_test)[:, 1] if hasattr(best_model, 'predict_proba') else None

        # Evaluate model performance on test data
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        mcc = matthews_corrcoef(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else np.nan

        # Append results to the list
        optimized_results.append({
            'Model': model_name,
            'Best_Params': grid_search.best_params_,
            'CV_Score': grid_search.best_score_,
            'Test_Accuracy': accuracy,
            'Test_Precision': precision,
            'Test_Recall': recall,
            'Test_F1': f1,
            'Test_MCC': mcc,
            'Test_ROC_AUC': roc_auc
        })

        # Print the best parameters and scores
        print(f"  Best params: {grid_search.best_params_}")
        print(f"  CV F1: {grid_search.best_score_:.4f}")
        print(f"  Test F1: {f1:.4f}")

    except Exception as e:
        # Catch and report any error during grid search
        print(f"  Error optimizing {model_name}: {e}")

# Convert results into a DataFrame
optimized_df = pd.DataFrame(optimized_results)

# Display the optimized results as a table (no index)
print("\nOptimized Model Results:\n")
print(optimized_df.round(4).to_string(index=False))
```
