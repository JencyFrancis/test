```python
#  Class Distribution Plot
def plot_class_distribution(data, fig, gs):
    """Plot class distribution pie chart"""

    # Create a subplot in the given figure and gridspec position
    ax1 = fig.add_subplot(gs[0, 0])

    # Count occurrences of each class in 'Machine failure' column
    machine_failure_counts = data['Machine failure'].value_counts()

    # Choose colors and labels
    colors = ['orange', 'yellow']
    labels = ['No Failure', 'Failure']

    # Generate the pie chart
    wedges, texts, autotexts = ax1.pie(machine_failure_counts.values,
                                       labels = None,
                                       autopct = '%.2f%%',
                                       colors = colors,
                                       startangle = 90,
                                       wedgeprops = {'alpha': 0.8})

    # Add legend
    ax1.legend(wedges, labels,
              title = "Failure Status",
              loc = "upper left",
              bbox_to_anchor = (1, 0, 0.5, 1))

    # Set the title
    ax1.set_title('Class Distribution', fontsize = 14, fontweight = 'bold')

# Create figure
fig = plt.figure(figsize = (12, 6))
gs = gridspec.GridSpec(1, 2)

# Call the function
plot_class_distribution(data, fig, gs)

# Adjust layout
plt.tight_layout()

# Display plot
plt.show()
```

##Statistical Significance Tests

```python
# Statistical steps for feature importance

# Initialize results list
results = []

# Chi-square tests (Type vs flags)
for flag in ['Machine_failure', 'TWF', 'HDF', 'PWF', 'OSF']:
    chi2, p = chi2_contingency(pd.crosstab(df1_scaled['Type_encoded'], df1_scaled[flag]))[:2]
    results.append({
        'Important Feature': flag,
        'Result': 'Significant association' if p < 0.05 else 'No association',
        'Test Type': 'Chi-square',
        'Test Statistic': f"{chi2:.3f}",
        'p-value': f"{p:.4f}"
    })

# T-tests (Numerical features vs failure)
for num_var in ['Air_temperature', 'Process_temperature', 'Rotational_speed', 'Torque', 'Tool_wear']:
    t_stat, p = ttest_ind(
        df1_scaled[df1_scaled['Machine_failure'] == 1][num_var],
        df1_scaled[df1_scaled['Machine_failure'] == 0][num_var],
        equal_var=False
    )
    results.append({
        'Important Feature': num_var,
        'Result': 'Significant difference' if p < 0.05 else 'No difference',
        'Test Type': 'T-test',
        'Test Statistic': f"{t_stat:.3f}",
        'p-value': f"{p:.4f}"
    })

# Create styled DataFrame
styled_df = (
    pd.DataFrame(results)
    .style
    .hide(axis="index")
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([
        {'selector': 'th, td', 'props': 'border: 1px solid black;'},
        {'selector': 'th', 'props': 'background-color: #f7f7f7;'}
    ])
)

# Display table
display(styled_df)
```

##Data Preparation

```python
# Select features for modeling
feature_columns = ['Air_temperature', 'Process_temperature', 'Rotational_speed',
                   'Torque', 'Tool_wear', 'Type_encoded']

# Define Feature Matrix(X) and Target(Y)
X = df1_scaled[feature_columns]
y = df1_scaled['Machine_failure']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 17, stratify = y
)

# Print the sizes and distribution
print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
print(f"Training set failure rate: {y_train.mean():.4f}")
print(f"Test set failure rate: {y_test.mean():.4f}")

# Convert to NumPy arrays
X_train_scaled = X_train.values
X_test_scaled = X_test.values
```

##Baseline Models

```python
print("\n" + "="*60)
print("BASELINE MODEL PERFORMANCE")
print("="*60)

# Initialize models
models = {
    'Random Forest': RandomForestClassifier(random_state = 17, n_estimators = 100),
    'MLP Classifier': MLPClassifier(random_state = 17, max_iter = 1000, early_stopping = True,
                                  validation_fraction = 0.2, n_iter_no_change = 10, verbose = True),
    'KNN': KNeighborsClassifier(n_neighbors = 5)
}

# Store baseline results
baseline_results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    start_time = time.time()

    # Use scaled data for MLP and KNN, original for Random Forest
    if name in ['MLP Classifier', 'KNN']:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1]

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average = 'binary')
    recall = recall_score(y_test, y_pred, average = 'binary')
    f1 = f1_score(y_test, y_pred, average = 'binary')
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    training_time = time.time() - start_time

    baseline_results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1_Score': f1,
        'ROC_AUC': roc_auc,
        'Training_Time': training_time
    }

    print(f"{name} - F1: {f1:.4f}, Recall: {recall:.4f}, ROC-AUC: {roc_auc:.4f}")

# Display baseline results table
print("\n=== BASELINE PERFORMANCE SUMMARY ===")
baseline_df = pd.DataFrame(baseline_results).T
print("+" + "-"*85 + "+")
print(f"| {'Model':<15} | {'Accuracy':<8} | {'Precision':<9} | {'Recall':<8} | {'F1-Score':<8} | {'ROC-AUC':<8} | {'Time(s)':<8} |")
print("+" + "-"*85 + "+")
for model_name, metrics in baseline_results.items():
    print(f"| {model_name:<15} | {metrics['Accuracy']:<8.4f} | {metrics['Precision']:<9.4f} | {metrics['Recall']:<8.4f} | {metrics['F1_Score']:<8.4f} | {metrics['ROC_AUC']:<8.4f} | {metrics['Training_Time']:<8.2f} |")
print("+" + "-"*85 + "+")
```

## Oversampling Technique

```python
# Initialize oversampling techniques
oversamplers = {
    'SMOTE': SMOTE(random_state=17),
    'ADASYN': ADASYN(random_state=17),
    'RandomOverSampler': RandomOverSampler(random_state=17),
    'SMOTETomek': SMOTETomek(random_state=17)
}

# Store oversampled results
oversampling_results = {}

for sampler_name, sampler in oversamplers.items():
    print(f"\n" + "="*50)
    print(f"APPLYING {sampler_name.upper()}")
    print("="*50)

    # Apply oversampling
    X_resampled, y_resampled = sampler.fit_resample(X_train, y_train)

    # Display class distribution after oversampling
    resampled_counts = pd.Series(y_resampled).value_counts().sort_index()
    print(f"\nClass distribution after {sampler_name}:")
    print(f"No Failure (0): {resampled_counts[0]} ({resampled_counts[0]/len(y_resampled)*100:.2f}%)")
    print(f"Failure (1): {resampled_counts[1]} ({resampled_counts[1]/len(y_resampled)*100:.2f}%)")

    # Scale resampled data for MLP and KNN
    X_resampled_scaled = scaler.fit_transform(X_resampled)

    sampler_results = {}

    # Train models on oversampled data
    for model_name, _ in models.items():
        print(f"\nTraining {model_name} with {sampler_name}...")
        start_time = time.time()

        # Create fresh model instance with enhanced MLP for better tracking
        if model_name == 'Random Forest':
            current_model = RandomForestClassifier(random_state=42, n_estimators=100)
            current_model.fit(X_resampled, y_resampled)
            y_pred = current_model.predict(X_test)
            y_pred_proba = current_model.predict_proba(X_test)[:, 1]
        elif model_name == 'MLP Classifier':
            # Enhanced MLP with better architecture and monitoring
            current_model = MLPClassifier(
                hidden_layer_sizes=(128, 64, 32),  # Multiple layers
                activation='relu',
                solver='adam',
                alpha=0.001,
                learning_rate_init=0.001,
                max_iter=1000,
                early_stopping=True,
                validation_fraction=0.2,
                n_iter_no_change=15,
                verbose=True,
                random_state=42
            )
            print(f"Training {model_name} with layers: (128, 64, 32)")
            current_model.fit(X_resampled_scaled, y_resampled)
            y_pred = current_model.predict(X_test_scaled)
            y_pred_proba = current_model.predict_proba(X_test_scaled)[:, 1]
            print(f"Converged after {current_model.n_iter_} iterations")
        else:  # KNN
            current_model = KNeighborsClassifier(n_neighbors=5)
            current_model.fit(X_resampled_scaled, y_resampled)
            y_pred = current_model.predict(X_test_scaled)
            y_pred_proba = current_model.predict_proba(X_test_scaled)[:, 1]

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary')
        recall = recall_score(y_test, y_pred, average='binary')
        f1 = f1_score(y_test, y_pred, average='binary')
        roc_auc = roc_auc_score(y_test, y_pred_proba)

        training_time = time.time() - start_time

        sampler_results[model_name] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1_Score': f1,
            'ROC_AUC': roc_auc,
            'Training_Time': training_time
        }

        print(f"{model_name} - F1: {f1:.4f}, Recall: {recall:.4f}, ROC-AUC: {roc_auc:.4f}")

    oversampling_results[sampler_name] = sampler_results
```

##Hyperparameter Optimization

```python
# Define parameter grids for each model
param_grids = {
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10]
    },
    'MLP Classifier': {
        'hidden_layer_sizes': [(64,), (128, 64), (128, 64, 32)],
        'learning_rate_init': [0.001, 0.01, 0.1],
        'alpha': [0.0001, 0.001, 0.01]
    },
    'KNN': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    }
}

# Find best oversampling technique for each model
best_samplers = {}
for model_name in models.keys():
    best_f1 = 0
    best_sampler = 'SMOTE'  # default
    for sampler_name in oversamplers.keys():
        current_f1 = oversampling_results[sampler_name][model_name]['F1_Score']
        if current_f1 > best_f1:
            best_f1 = current_f1
            best_sampler = sampler_name
    best_samplers[model_name] = best_sampler
    print(f"Best sampler for {model_name}: {best_sampler} (F1: {best_f1:.4f})")

# Hyperparameter optimization results
optimization_results = {}

for model_name in models.keys():
    print(f"\n" + "="*40)
    print(f"OPTIMIZING {model_name.upper()}")
    print("="*40)

    # Get best oversampling data for this model
    best_sampler = best_samplers[model_name]
    X_resampled, y_resampled = oversamplers[best_sampler].fit_resample(X_train, y_train)

    # Scale if needed
    if model_name in ['MLP Classifier', 'KNN']:
        X_resampled_scaled = scaler.fit_transform(X_resampled)
        X_opt = X_resampled_scaled
        X_test_opt = X_test_scaled
    else:
        X_opt = X_resampled
        X_test_opt = X_test

    # Create model instance
    if model_name == 'Random Forest':
        base_model = RandomForestClassifier(random_state=42)
    elif model_name == 'MLP Classifier':
        base_model = MLPClassifier(random_state=42, max_iter=1000, early_stopping=True)
    else:
        base_model = KNeighborsClassifier()

    # Perform grid search
    grid_search = GridSearchCV(
        base_model,
        param_grids[model_name],
        cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
        scoring='f1',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_opt, y_resampled)

    # Evaluate best model
    best_model = grid_search.best_estimator_
    y_pred_opt = best_model.predict(X_test_opt)
    y_pred_proba_opt = best_model.predict_proba(X_test_opt)[:, 1]

    # Calculate optimized metrics
    opt_metrics = {
        'Best_Params': grid_search.best_params_,
        'Best_CV_F1': grid_search.best_score_,
        'Test_Accuracy': accuracy_score(y_test, y_pred_opt),
        'Test_Precision': precision_score(y_test, y_pred_opt),
        'Test_Recall': recall_score(y_test, y_pred_opt),
        'Test_F1': f1_score(y_test, y_pred_opt),
        'Test_ROC_AUC': roc_auc_score(y_test, y_pred_proba_opt),
        'Best_Sampler': best_sampler
    }

    optimization_results[model_name] = opt_metrics

    print(f"Best Parameters: {grid_search.best_params_}")
    print(f"Best CV F1-Score: {grid_search.best_score_:.4f}")
    print(f"Test F1-Score: {opt_metrics['Test_F1']:.4f}")
    print(f"Test Recall: {opt_metrics['Test_Recall']:.4f}")
    print(f"Test ROC-AUC: {opt_metrics['Test_ROC_AUC']:.4f}")

# Print optimization summary table
print("\n=== HYPERPARAMETER OPTIMIZATION RESULTS ===")
print("+" + "-"*95 + "+")
print(f"| {'Model':<15} | {'Best Sampler':<15} | {'CV F1':<8} | {'Test F1':<8} | {'Test Recall':<11} | {'Test ROC-AUC':<12} |")
print("+" + "-"*95 + "+")
for model_name, metrics in optimization_results.items():
    print(f"| {model_name:<15} | {metrics['Best_Sampler']:<15} | {metrics['Best_CV_F1']:<8.4f} | {metrics['Test_F1']:<8.4f} | {metrics['Test_Recall']:<11.4f} | {metrics['Test_ROC_AUC']:<12.4f} |")
print("+" + "-"*95 + "+")
```

```python
print("\n" + "="*60)
print("LEARNING CURVES ANALYSIS")
print("="*60)

# Create learning curves for optimized models
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for idx, (model_name, opt_result) in enumerate(optimization_results.items()):
    print(f"\nGenerating learning curve for {model_name}...")

    # Get optimized model and data
    if model_name == 'Random Forest':
        model = RandomForestClassifier(**opt_result['Best_Params'], random_state=42)
        X_data = X_train
        y_data = y_train
    elif model_name == 'MLP Classifier':
        model = MLPClassifier(**opt_result['Best_Params'], random_state=42, max_iter=1000)
        X_data = X_train_scaled
        y_data = y_train
    else:
        model = KNeighborsClassifier(**opt_result['Best_Params'])
        X_data = X_train_scaled
        y_data = y_train

    # Generate learning curves
    train_sizes, train_scores, val_scores = learning_curve(
        model, X_data, y_data,
        cv=5,
        n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 10),
        scoring='f1'
    )

    # Calculate mean and std
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)

    # Plot learning curves
    ax = axes[idx]
    ax.plot(train_sizes, train_mean, 'o-', color='blue', label='Training F1-Score')
    ax.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    ax.plot(train_sizes, val_mean, 'o-', color='red', label='Validation F1-Score')
    ax.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='red')

    ax.set_title(f'{model_name}\nLearning Curve', fontweight='bold', fontsize=12)
    ax.set_xlabel('Training Set Size', fontweight='bold')
    ax.set_ylabel('F1-Score', fontweight='bold')
    ax.legend(loc='lower right')
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 1.1)

plt.tight_layout()
plt.show()
```
